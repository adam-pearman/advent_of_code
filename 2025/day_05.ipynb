# 2025 - Day 5

```python
with open('input.txt') as f:
  lines = [line.strip() for line in f]

ranges, ids = [], []
done = False

for line in lines:
  if line == '':
    done = True
    continue
  if done:
    ids.append(int(line))
  else:
    ranges.append(tuple(map(int, line.split('-'))))

ranges.sort()
ids.sort()
```

## Part 1

```python
fresh = set()
i = 0

for lo, hi in ranges:
  while i < len(ids) and ids[i] < lo:
    i += 1

  j = i
  while j < len(ids) and ids[j] <= hi:
    fresh.add(ids[j])
    j += 1

print(len(fresh))
```

## Part 2

```python
lo, hi = ranges[0]
total = hi - lo + 1

for clo, chi in ranges[1:]:
  if clo <= hi:
    if chi > hi:
      total += chi - hi
      hi = chi
  else:
    total += chi - clo + 1
    lo, hi = clo, chi

print(total)
```
