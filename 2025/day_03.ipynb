#Part 1

```python
with open('input.txt') as f:
  lines = [line.strip() for line in f]
```

```python
part1 = 0

for line in lines:
  best1 = ''
  best2 = ''

  for i in range(len(line)):
    ch = line[i]
    if ch > best1 and i < len(line) - 1:
      best1 = ch
      best2 = ''
    elif ch > best2:
      best2 = ch

  part1 += int(best1 + best2)

print(part1)
```

#Part 2

```python
part2 = 0

for line in lines:
  stack = []
  drop = len(line) - 12

  for ch in line:
    while drop and stack and stack[-1] < ch:
      stack.pop()
      drop -= 1
    stack.append(ch)

  part2 += int(''.join(stack[:12]))

print(part2)
```
