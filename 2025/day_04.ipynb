#Part 1

```python
with open('input.txt') as f:
  grid = [list(map(str,line.strip())) for line in f]

rows = len(grid)
cols = len(grid[0])

def can_access(r, c, remove = False):
  cell = grid[r][c]

  if cell == '.':
    return False

  n = 0

  # UL
  if r > 0 and c > 0 and grid[r-1][c-1] == '@':
    n += 1

  # U
  if r > 0 and grid[r-1][c] == '@':
    n += 1

  # UR
  if r > 0 and c < cols - 1 and grid[r-1][c+1] == '@':
    n += 1

  # L
  if c > 0 and grid[r][c-1] == '@':
    n += 1

  # R
  if c < cols - 1 and grid[r][c+1] == '@':
    n += 1

  # DL
  if r < rows - 1 and c > 0 and grid[r+1][c-1] == '@':
    n += 1

  # D
  if r < rows - 1 and grid[r+1][c] == '@':
    n += 1

  # DR
  if r < rows - 1 and c < cols - 1 and grid[r+1][c+1] == '@':
    n += 1

  if n < 4 and remove:
    grid[r][c] = '.'

  return n < 4
```

```python
print(sum(int(can_access(r, c)) for r in range(rows) for c in range(cols)))
```

#Part 2

```python
total = 0

while True:
  count = sum(int(can_access(r, c, True)) for r in range(rows) for c in range(cols))

  if count == 0:
    break

  total += count

print(total)
```
