#Part 1

```python
with open('input.txt') as f:
  lines = [line.strip() for line in f]
```

```python
dial = 50
ans = 0

for line in lines:
  direction = line[0]
  amount = int(line[1:])

  if direction == 'L':
    dial -= amount
  else:
    dial += amount

  dial %= 100

  if dial == 0:
    ans += 1
print(ans)
```

#Part 2

```python
dial = 50
ans = 0

for line in lines:
  direction = line[0]
  amount = int(line[1:])
  delta = amount if direction == 'R' else -amount

  if delta > 0:
    k = (100 - dial) % 100
    if k == 0:
      k = 100
    if delta >= k:
      ans += 1 + (delta - k) // 100
  elif delta < 0:
    k = dial % 100
    if k == 0:
      k = 100
    if -delta >= k:
      ans += 1 + (-delta - k) // 100

  dial = (dial + delta) % 100

print(ans)
```
