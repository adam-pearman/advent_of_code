# 2025 - Day 11

```python
from functools import cache

graph = {}

with open('input.txt') as f:
  for line in f:
    vertex, neighbours = line.split(': ')
    graph[vertex] = neighbours.split()
```

## Part 1

```python
@cache
def part1(vertex):
    if vertex == "out":
        return 1
    return sum(part1(neighbour) for neighbour in graph[vertex])

print(part1("you"))
```

## Part 2

```python
@cache
def part2(vertex, seen_fft=False, seen_dac=False):
    if vertex == "fft":
        seen_fft = True
    if vertex == "dac":
        seen_dac = True

    if vertex == "out":
        return 1 if seen_fft and seen_dac else 0

    return sum(
        part2(neighbour, seen_fft, seen_dac)
        for neighbour in graph[vertex]
    )

print(part2("svr"))
```
