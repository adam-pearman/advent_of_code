# 2025 - Day 6

```python
from math import prod

def calculate(problem):
  nums = list(map(int, problem[:-1]))
  return sum(nums) if problem[-1] == '+' else prod(nums)
```

## Part 1

```python
with open('input.txt') as f:
  cols = list(zip(*[line.split() for line in f]))

print(sum(calculate(col) for col in cols))
```

## Part 2

```python
with open('input.txt') as f:
  cols = list(zip(*[line for line in f]))

problems = []

i = 0
while i < len(cols):
  operator = ''
  problem = []
  k = 0
  for j, chars in enumerate(cols[i:]):
    if all(ch == ' ' for ch in chars):
      k = j + 1
      break
    if chars[-1] != ' ':
      operator = chars[-1]
    problem.append(int(''.join(chars[:-1])))
  problem.append(operator)
  problems.append(problem)
  if k == 0:
    break
  i += k

print(sum(calculate(problem) for problem in problems))
```
