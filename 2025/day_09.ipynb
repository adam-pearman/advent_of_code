# 2025 - Day 9

```python
with open('input.txt') as f:
  tiles = [tuple(map(int, line.split(','))) for line in f]

n = len(tiles)

def get_area(tile1, tile2):
  c1, r1 = tile1
  c2, r2 = tile2
  return (abs(c1 - c2) + 1) * (abs(r1 - r2) + 1)

def is_outside(p0_x, p1_x, lo_x, hi_x):
  return (p0_x <= lo_x and p1_x <= lo_x) or (p0_x >= hi_x and p1_x >= hi_x)
```

## Part 1

```python
max_area = 0

for i, t1 in enumerate(tiles):
  for t2 in tiles[i+1:]:
    max_area = max(max_area, get_area(t1, t2))

print(max_area)
```

## Part 2

```python
max_area = 0

pairs = [(a, b) for i, a in enumerate(tiles) for b in tiles[i+1:]]
edges = [(tile, tiles[i-1]) for i, tile in enumerate(tiles)]

for a, b in pairs:
  lo = [min(a[0], b[0]), min(a[1], b[1])]
  hi = [max(a[0], b[0]), max(a[1], b[1])]

  if all(
      is_outside(p0[0], p1[0], lo[0], hi[0]) or
      is_outside(p0[1], p1[1], lo[1], hi[1])
      for p0, p1 in edges
    ):
    max_area = max(max_area, get_area(lo, hi))

print(max_area)
```
