# 2025 - Day 12

```python
with open("input.txt") as f:
    lines = [line.strip() for line in f]

presents = []
regions = []

i = 0
n = len(lines)

while i < n:
  line = lines[i]

  if line.endswith(":"):
    shape = []
    i += 1

    while i < n and lines[i] and set(lines[i]).issubset({"#", "."}):
      shape.append(list(lines[i].strip()))
      i += 1

    size = sum(row.count('#') for row in shape)
    presents.append({"shape": shape, "size": size})
    continue

  if "x" in line and ":" in line:
    name, nums = line.split(":")
    required = list(map(int, nums.split()))
    regions.append((name, required))

  i += 1
```

```python
total = 0

for name, required in regions:
  w, h = map(int, name.split('x'))
  area = w * h

  for idx, count in enumerate(required):
    if count:
      area -= presents[idx]['size'] * count

  if area >= 0:
    total += 1

print(total)
```
