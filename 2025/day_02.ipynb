#Part 1

```python
with open('input.txt') as f:
  ranges = f.read().split(',')
```

```python
def part_1(lo, hi):
  results = []

  for length in range (len(str(lo)), len(str(hi)) + 1):
    if length % 2:
      continue

    half = length // 2

    for h in range(10**(half - 1), 10**half):
      result = int(str(h) + str(h))
      if lo <= result <= hi:
        results.append(result)

  return results

ans = 0

for r in ranges:
  lo, hi = map(int, r.split('-'))
  ans += sum(part_1(lo, hi))

print(ans)
```

#Part 2

```python
def part_2(lo, hi):
  results = set()

  for length in range(len(str(lo)), len(str(hi)) + 1):
    for block_size in range(1, length):
      if length % block_size != 0:
        continue

      repeats = length // block_size
      if repeats < 2:
        continue

      for block in range(10**(block_size - 1), 10**block_size):
        result = int(str(block) * repeats)
        if lo <= result <= hi:
          results.add(result)

  return results

ans = 0

for r in ranges:
  lo, hi = map(int, r.split('-'))
  ans += sum(part_2(lo, hi))

print(ans)
```
