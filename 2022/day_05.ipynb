# 2022 - Day 5

```python
import re

start = []
instructions = []
with open('input.txt') as input:
  init_complete = False
  for line in input:
    line = line.rstrip()
    if line == '':
      init_complete = True
      continue
    if init_complete:
      instructions.append(re.findall(r'\d+', line))
    else:
      start.append(line)
```

## Part 1

```python
stacks = {}
for i in range(len(start)):
  line = start[len(start) - 1 - i]
  if i == 0:
    for j in line.rsplit():
      stacks[j] = []
  else:
    for j in range(len(line)):
      if line[j] == '[':
        stacks[str(j // 4 + 1)].append(line[j+1])

for instruction in instructions:
  for i in range(int(instruction[0])):
    crate = stacks[instruction[1]].pop()
    stacks[instruction[2]].append(crate)

ans = ''
for stack in stacks:
  ans += stacks[stack].pop()
print(ans)
```

## Part 2

```python
stacks = {}
for i in range(len(start)):
  line = start[len(start) - 1 - i]
  if i == 0:
    for j in line.rsplit():
      stacks[j] = []
  else:
    for j in range(len(line)):
      if line[j] == '[':
        stacks[str(j // 4 + 1)].append(line[j+1])

for instruction in instructions:
  crates = []
  for i in range(int(instruction[0])):
    crates.append(stacks[instruction[1]].pop())
  stacks[instruction[2]] += crates[::-1]

ans = ''
for stack in stacks:
  ans += stacks[stack].pop()
print(ans)
```
