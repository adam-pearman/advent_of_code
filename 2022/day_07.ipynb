# 2022 - Day 7

```python
with open('input.txt') as f:
  lines = [line.strip() for line in f]
```

```python
from collections import defaultdict

sizes = defaultdict(int)
cwd = []

for line in lines:
  if line.startswith('$ cd'):
    dest = line.split()[-1]
    if dest == '/':
      cwd = []
    elif dest == '..':
      cwd.pop()
    else:
      cwd.append(dest)
  elif line[0].isdigit():
    size, name = line.split()
    size = int(size)

    for i in range(len(cwd) + 1):
      path = '/' + '/'.join(cwd[:i])
      sizes[path] += size
```

## Part 1

```python
ans = 0
for _, size in sizes.items():
  if size <= 100000:
    ans += size

print(ans)
```

## Part 2

```python
total_space = 70000000
required_space = 30000000
unused_space = total_space - sizes['/']

ans = total_space

for _, size in sizes.items():
  potential_space = unused_space + size
  if potential_space >= required_space:
    ans = min(ans, size)

print(ans)
```
