# 2022 - Day 2

## Part 1

```python
scores = {
    'A': {
        'X': 4,
        'Y': 8,
        'Z': 3,
    },
    'B': {
        'X': 1,
        'Y': 5,
        'Z': 9,
    },
    'C': {
        'X': 7,
        'Y': 2,
        'Z': 6,
    },
}
```

```python
with open('input.txt') as input:
  total = 0
  for line in input:
    x, y = line.rsplit()
    total += scores[x][y]
  print(total)
```

## Part 2

```python
outcomes = {
    'A': {
        'X': 'Z',
        'Y': 'X',
        'Z': 'Y',
    },
    'B': {
        'X': 'X',
        'Y': 'Y',
        'Z': 'Z',
    },
    'C': {
        'X': 'Y',
        'Y': 'Z',
        'Z': 'X',
    },
}
```

```python
with open('input.txt') as input:
  total = 0
  for line in input:
    x, y = line.rsplit()
    y = outcomes[x][y]
    total += scores[x][y]
  print(total)
```
