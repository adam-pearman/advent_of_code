# 2022 - Day 1

```python
from heapq import heapify, heappop

totals = []

with open('input.txt') as input:
  total = 0
  for line in input:
    line = line.rstrip()
    if line == '':
      totals.append(total)
      total = 0
    else:
      total += int(line)
```

## Part 1

```python
nums = [-x for x in totals]
heapify(nums)
print(-heappop(nums))
```

## Part 2

```python
nums = [-x for x in totals]
heapify(nums)
print(sum([-heappop(nums) for _ in range(3)]))
```
