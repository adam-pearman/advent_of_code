#Part 1

```python
with open('input.txt') as input:
  total = 0
  for line in input:
    x, y = line.rstrip().split(',')
    x, y = x.split('-'), y.split('-')
    x, y = [int(i) for i in x], [int(i) for i in y]
    if (y[0] <= x[0] <= y[1] and y[0] <= x[1] <= y[1]) or (x[0] <= y[0] <= x[1] and x[0] <= y[1] <= x[1]):
      total += 1
  print(total)
```
