# 2022 - Day 8

```python
with open('input.txt') as f:
  grid = [list(map(int, line.strip())) for line in f]

rows = len(grid)
cols = len(grid[0])
```

## Part 1

```python
def visible(r, c):
  h = grid[r][c]

  if all(grid[r][x] < h for x in range(0, c)):
    return True

  if all(grid[r][x] < h for x in range(c+1, cols)):
    return True

  if all(grid[y][c] < h for y in range(0, r)):
    return True

  if all(grid[y][c] < h for y in range(r+1, rows)):
    return True

  return False

print(sum(visible(r, c) for r in range(rows) for c in range(cols)))
```

## Part 2

```python
def scenic_score(r, c):
  h = grid[r][c]

  left = 0
  for x in range(c-1, -1, -1):
    left += 1
    if grid[r][x] >= h:
      break

  right = 0
  for x in range(c+1, cols):
    right += 1
    if grid[r][x] >= h:
      break

  up = 0
  for y in range(r-1, -1, -1):
    up += 1
    if grid[y][c] >= h:
      break

  down = 0
  for y in range(r+1, rows):
    down += 1
    if grid[y][c] >= h:
      break

  return left * right * up * down

print(max(scenic_score(r, c) for r in range(rows) for c in range(cols)))
```
