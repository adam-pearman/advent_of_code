#Part 1

```python
with open('input.txt') as input:
  total = 0
  for line in input:
    line = line.rstrip()
    mid = len(line) // 2
    l, r = set(line[:mid]), set(line[mid:])
    ans = (ord((l & r).pop())) - 96
    if ans < 0:
      ans += 58
    total += ans
  print(total)
```

#Part 2

```python
with open('input.txt') as input:
  total = 0
  group = []
  for line in input:
    line = line.rstrip()
    group.append(set(line))
    if len(group) == 3:
      ans = ord((group[0] & group[1] & group[2]).pop()) - 96
      if ans < 0:
        ans += 58
      total += ans
      group = []
  print(total)
```
