# 2022 - Day 6

```python
with open('input.txt') as f:
  input = f.read()

def find_marker(k):
  window = input[0:k]

  for right in range(k, len(input)):
    if len(set(window)) == k:
      return right - 1
    left = right - k
    window = input[left:right]

  return -1
```

## Part 1

```python
print(find_marker(4))
```

## Part 2

```python
print(find_marker(14))
```
