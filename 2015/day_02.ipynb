# Day 2

```python
with open('input.txt') as f:
  presents = [tuple(map(int, line.split('x'))) for line in f]
```

# Part 1

```python
total = 0

for l, w, h in presents:
  lw, wh, hl = l * w, w * h, h * l
  total += (lw + wh + hl) * 2 + min(lw, wh, hl)

print(total)
```

# Part 2

```python
total = 0

for l, w, h in presents:
  present = sorted([l, w, h])
  total += sum(present[:-1]) * 2 + l * w * h

print(total)
```
