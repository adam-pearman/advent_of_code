# 2015 - Day 3

```python
with open('input.txt') as f:
  directions = f.read()

def move(position, direction):
  r, c = position

  match direction:
    case '^':
      r -= 1
    case '>':
      c += 1
    case 'v':
      r += 1
    case '<':
      c -= 1

  return (r, c)
```

## Part 1

```python
current = (0, 0)
seen = set([current])

for direction in directions:
  current = move(current, direction)
  seen.add(current)

print(len(seen))
```

## Part 2

```python
santa, robo = (0, 0), (0, 0)
seen = set([santa])

for i, direction in enumerate(directions):
  if i % 2 == 0:
    santa = move(santa, direction)
    seen.add(santa)
  else:
    robo = move(robo, direction)
    seen.add(robo)

print(len(seen))
```
