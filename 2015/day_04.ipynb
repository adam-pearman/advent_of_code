# 2015 - Day 4

```python
from hashlib import md5

with open('input.txt') as f:
  input = f.read()
```

## Part 1

```python
i = 0
while True:
  hash = md5(f'{input}{i}'.encode()).hexdigest()
  if hash.startswith('00000'):
    print(i)
    break
  i += 1
```
