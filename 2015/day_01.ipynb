# Day 1

```python
with open('input.txt') as f:
  input = f.read()
```

# Part 1

```python
count = 0

for ch in input:
  if ch == '(':
    count += 1
  else:
    count -= 1

print(count)
```

# Part 2

```python
count = 0

for i, ch in enumerate(input):
  if ch == '(':
    count += 1
  else:
    count -= 1
  if count == -1:
    print(i + 1)
    break
```
