# 2025 - Day 7

```python
with open('input.txt') as f:
  grid = [list(map(str, line.strip())) for line in f]

start = ()

for c in range(len(grid[0])):
  if grid[0][c] == 'S':
    start = (0, c)
    break
```

## Part 1

```python
stack = [start]
splits = set()

while stack:
  r, c = stack.pop()
  while r < len(grid) - 1:
    r += 1
    if (r, c) in splits:
      break
    if grid[r][c] == '^':
      splits.add((r, c))
      stack.append((r, c - 1))
      stack.append((r, c + 1))
      break

print(len(splits))
```

## Part 2

```python
from collections import defaultdict

rows = len(grid)
cols = len(grid[0])

paths = defaultdict(int)
paths[start] = 1

for r in range(rows):
  next_paths = defaultdict(int)
  for (rr, c), p in list(paths.items()):
    if rr != r:
      continue
    nr = r + 1
    if nr >= rows:
      next_paths[(nr, c)] += p
      continue
    if grid[nr][c] == '^':
      next_paths[(nr, c - 1)] += p
      next_paths[(nr, c + 1)] += p
    else:
      next_paths[(nr, c)] += p
  paths = next_paths

print(sum(p for (r, c), p in paths.items() if r == rows))
```
